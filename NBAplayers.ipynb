```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import colorsys
from random import *
import os
print(os.getcwd())
```

```python
#LeBron James DataFrame
loc='project/lebron_stats.csv'
df=pd.read_csv(loc)
df
df.drop([16,17,18,19,20])
```

```python
#Kobe Bryant DataFrame
df2=pd.read_csv('project/kobe_bryant_stats.csv')
df2
df2.drop([20,])
```

```python
#Michael Jordan DataFrame
loc3='project/michael_jordan_stats.csv'
df3=pd.read_csv(loc3)
df3
df3.drop([9,14,15,16,19,20,21,22]) #Rows dropped because of retirement
```

```python
#LeBron vs. Kobe scatter plot
'''
Data Cleaning:
Dropping indices using DataFrame.drop
Slice method used to strip the String for Season into a full-year
Pandas.to_numeric used to convert the full-year into a integer so that 
it's accepted as axes variables.
'''

new_df=df.drop([13,14,15,16,17,18,19,20]) #Dropping data indices where the players were
new_df2=df2.drop([0,1,2,3,4,5,6,20])      #not in the league at the same time.

x=new_df['Season']
y=new_df['PTS']
x=x.str.slice(0,4)
pd.to_numeric(x)

x2=new_df2['Season']
y2=new_df2['PTS']
x2=x2.str.slice(0,4)
pd.to_numeric(x2)

plt.title('King James vs. The Black Mamba')
plt.scatter(x,y,marker='o', color='red')
plt.scatter(x2,y2,marker='o', color='black')
plt.xlabel('Season')
plt.ylabel('Average Points per game')
plt.tight_layout(-10) #Spaces out x-axis data so that it no longer overlaps

plt.show()
```

```python
#Kobe's average point statistics over the course of his career
new_df2=df2.drop([20,])
x2=new_df2['Season']
x2=x2.str.slice(0,4)
pd.to_numeric(x2)
y2=new_df2['PTS']
s = pd.to_numeric(x2) # converts the 4-char year strings to numbers.
plt.scatter(x2,y2,marker='o',color='purple')
plt.title('Kobe Bryant Career Points')
plt.xlabel('Season')
plt.ylabel('Points per game')
plt.tight_layout(-10) #Spaces out x-axis data so that it no longer overlaps

plt.show()
```
